```python
class Tier:
    def __init__(self, name, alter):
        self.name = name
        self.alter = alter
        self.hunger = 50 # 0-100

    def geraeusch_machen(self):
        pass
    
    def fuettern(self):
        self.hunger = max(0, self.hunger - 20)
        return f"{self.name} wurde gefüttert. Hunger: {self.hunger}"
    
    def spielen(self):
        self.hunger = min(100, self.hunger + 10)
        return f"{self.name} spielt. Hunger: {self.hunger}"

    def ist_hungrig(self):
        return self.hunger > 70 # Beispiel: Wenn Hunger über 70, ist das Tier hungrig
    
    def ist_satt(self):        return self.hunger < 30 # Beispiel: Wenn Hunger unter 30, ist das Tier satt  

    
class Hund(Tier):   
    def geraeusch_machen(self):
        return "Wuff!"
    def spielen(self):
        self.hunger = min(100, self.hunger + 15) # Hunde verbrauchen mehr Energie beim Spielen
        return f"{self.name} spielt und bellt. Hunger: {self.hunger}"
    
class Katze(Tier):
    def geraeusch_machen(self):
        return "Miau!"
    def spielen(self):
        self.hunger = min(100, self.hunger + 30)  
        if self.hunger > 80:
            return f"{self.name} faucht! Sie ist zu hungrig zum Spielen. Hunger: {self.hunger}"
        return f"{self.name} spielt kurz und ignoriert dich danach wieder. Hunger: {self.hunger}"
    
class Papagai(Tier):
    def geraeusch_machen(self):
        return "Halllo, Hallo!"
    def spielen(self):
        self.hunger = min(100, self.hunger + 5)  
        return f"{self.name} flattert herum und plappert begeistert. Hunger: {self.hunger}"
    
def menue_anzeigen():
    print("\n--- Tier-Interaktion Menue ---")
    print("1. Tier füttern")
    print("2. Tier spielen lassen")
    print("3. Tiergeräusch machen")
    print("4. Info anzeigen") 
    print("m. Menue anzeigen")
    print("0. Beenden")

def menue(tier):
    menue_anzeigen()
    while True:
        auswahl = input("Wähle eine Option: ").strip().lower()
        if auswahl == "1":
            if tier.ist_satt():
                print(f"{tier.name} ist jetzt satt und zufrieden!")
                continue
            print(">>", tier.fuettern())
        elif auswahl == "2":
            if tier.ist_hungrig():
                print(f"{tier.name} ist zu hungrig zum Spielen! Bitte füttere es zuerst.")
                continue
            print(">>", tier.spielen())
        elif auswahl == "3":
            print(">>", tier.geraeusch_machen())
        elif auswahl == "4":
            print(f"Name: {tier.name}, Alter: {tier.alter}, Hunger: {tier.hunger}")
        elif auswahl.lower() == "m":
            menue_anzeigen()
        elif auswahl == "0":
            print("Programm wird beendet.")
            break
        else:
            print("Ungültige Auswahl. Bitte versuche es erneut.")
if __name__ == "__main__":
    print("Willkommen zum Tier-Interaktionsprogramm!")
    tier_auswahl = input("Wähle ein Tier (Hund, Katze, Papagai): ").strip().lower()
    
    if tier_auswahl == "hund":
        tier = Hund(name="Bello", alter=3)
    elif tier_auswahl == "katze":
        tier = Katze(name="Mia", alter=2)
    elif tier_auswahl == "papagai":
        tier = Papagai(name="Polly", alter=5)
    else:
        print("Ungültige Tierauswahl. Standardmäßig wird ein Hund erstellt.")
        tier = Hund(name="Bello", alter=3)
    
    menue(tier)

```
